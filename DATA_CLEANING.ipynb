```python
#To check for duplicate nurmbers
Dirty_sales.drop_duplicates()
```

```python
#TO Drop rows where Order_ID or Order_Date is missing (critical fields)
Dirty_sales = Dirty_sales.dropna(subset=['Order_ID', 'Order_Date'])
```

```python
print(Dirty_sales.isnull().sum())
```

```python
Dirty_sales.head()
```

```python
#To Drop rows with missing Customer_ID if it's not needed for analysis
# Otherwise, fill with placeholder like "Unknown" or 0
Dirty_sales['Customer_ID'] = Dirty_sales['Customer_ID'].fillna('Unknown')
```

```python
Dirty_sales.head()
```

```python
#  TO Fill missing countries with a placeholder or mode (most common country)
Dirty_sales['Country'] = Dirty_sales['Country'].fillna(Dirty_sales['Country'].mode()[0])
```

```python
Dirty_sales.head()
```

```python
# To Replace text numbers with real numbers
text_to_num = {
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5
}

for col in ['Quantity', 'Unit_Price']:
    Dirty_sales[col] = Dirty_sales[col].replace(text_to_num)

# Convert to numeric and handle errors
Dirty_sales['Quantity'] = pd.to_numeric(Dirty_sales['Quantity'], errors='coerce')
Dirty_sales['Unit_Price'] = pd.to_numeric(Dirty_sales['Unit_Price'], errors='coerce')
Dirty_sales['Total_Amount'] = pd.to_numeric(Dirty_sales['Total_Amount'], errors='coerce')

#  Remove negative values
Dirty_sales.loc[Dirty_sales['Quantity'] < 0, 'Quantity'] = np.nan
Dirty_sales.loc[Dirty_sales['Unit_Price'] < 0, 'Unit_Price'] = np.nan
Dirty_sales.loc[Dirty_sales['Total_Amount'] < 0, 'Total_Amount'] = np.nan

#  Fill Quantity and Unit_Price missing values with median
Dirty_sales['Quantity'] = Dirty_sales['Quantity'].fillna(Dirty_sales['Quantity'].median())
Dirty_sales['Unit_Price'] = Dirty_sales['Unit_Price'].fillna(Dirty_sales['Unit_Price'].median())

#  To Recalculate Total_Amount for all rows
Dirty_sales['Total_Amount'] = Dirty_sales['Quantity'] * Dirty_sales['Unit_Price']

```
